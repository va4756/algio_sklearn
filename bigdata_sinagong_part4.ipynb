### 문2-1) 방문객 합계와 관광개의 비율을 계산하시오.

## II.작업형2 유형

### # 문제정의
#### 평가: MAE
### target: TotalCharges
### 최종파일: result.csv(컬럼 1개 pred)

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd
# train = pd.read_csv("churn_train.csv")
# test = pd.read_csv("churn_test.csv")
train = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/churn_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/churn_test.csv")

# 탐색적 데이터 분석(EDA)
print(train.shape, test.shape)
print(train.info())
print(train.isnull().sum().sum())
print(train['customerID'].nunique())
```

```python
print(train.shape, test.shape)
```

```python
# print(train['TotalCharges'].describe())
print(train.describe())
```

```python
# 데이터 전처리
train = train.drop(['customerID'], axis=1)
test = test.drop(['customerID'], axis=1)
target = train.pop('TotalCharges')
print(train.shape, test.shape, target.shape)
```

```python
# 레이블 인코딩
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

cols = train.select_dtypes(include='object').columns

for col in cols:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.fit_transform(test[col])

# 검증데이터 분리
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=123)

# 랜덤포레스트 -> 회귀모델 사용
rfr = RandomForestRegressor(random_state=123)
rfr.fit(X_tr, y_tr)
pred = rfr.predict(X_val)

# MAE
res_rfr = mean_absolute_error(y_val, pred)
print(res_rfr)
```

```python
# LightGBM 모델 사용
import lightgbm as lgb
from sklearn.metrics import mean_absolute_error

lg = lgb.LGBMRegressor(random_state=123, verbose=-1)
lg.fit(X_tr, y_tr)
pred = lg.predict(X_val)
res_lg = mean_absolute_error(y_val, pred)
print(res_lg)
```

```python
# 최종 제출 파일
pred = rfr.predict(test)
result = pd.DataFrame({'pred': pred})
result.to_csv('result.csv', index=False)
```

```python
# 1. pred 행의 수
print(pred.shape)

# 2. 생성한 csv 확인
df = pd.read_csv('result.csv', index_col=0)
df.head()
```

## III.작업형3 유형

### 문1-1)

```python
import pandas as pd
# df = pd.read_csv("churn.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/churn.csv")
print(df.shape)
df.head()
```

```python
df.info()
```

```python
# from statsmodels.formula.api import logit
# # 1) 로지스틱 회귀분석을 위한 포뮬라 생성
# formula = "Churn ~ AccountWeeks + ContractRenewal + DataPlan + DataUsage + CustServCalls + DayMins + DayCalls + MonthlyCharge + OverageFee + RoamMins"
```

```python
# 변수를 리스트로 만들고, +로 결합하는 방식입니다.
from statsmodels.formula.api import logit
features = [
    "AccountWeeks", "ContractRenewal", "DataPlan", "DataUsage",
    "CustServCalls", "DayMins", "DayCalls", "MonthlyCharge",
    "OverageFee", "RoamMins"
]
formula = "Churn ~ " + " + ".join(features)
model = logit(formula, data=df).fit()

# 3) 유의하지 않은 독립변수의 개수 구하기
print(model.summary())
```

```python
sum(model.pvalues[1:] > 0.05)  # Intercept는 독립변수가 아니므로 제외
```

### 문1-2)

```python
# 1) 유의한 변수 선택
print(model.pvalues < 0.05)
```

```python
from statsmodels.formula.api import logit
# 1) 유의한 변수 선택
formula = "Churn ~ DataUsage + DayMins"

# 2) 로지스틱 회귀
model = logit(formula, data=df).fit()

# 3) 유의한 회귀계수 합계
print(model.summary())
```
