### 문1-1) 대륙별 맥주 소비량의 평균을 구하여, 가장 많이 소비하는 대륙을 찾으시오.

```python
import pandas as pd
# df = pd.read_csv("drinks.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/drinks.csv")
print(df.shape)
df.head()
```

```python
df.info()
```

```python
continent = df.groupby('continent')['beer_servings'].mean()
continent
```

```python
top_continent = continent.idxmax()
top_continent
```

### 문1-2) 최대소비 대륙에서의 국가별 맥주 소비량을 구하시오(5번째 국가의 소비량을 찾으시오).

```python
cond = df['continent'] == top_continent
df = df[cond]
df = df.sort_values('beer_servings', ascending=False)
df
#df.iloc[4]['beer_servings']
```

### 문2-1) 방문객 합계와 관광개의 비율을 계산하시오.

```python
import pandas as pd
# df = pd.read_csv("tourist.csv")
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch8/tourist.csv")
print(df.shape)
df.head()
```

```python
df['방문객합계'] = df[['관광', '공무', '사업', '기타']].sum(axis=1)
df['관광객비율'] = df['관광']/ df['방문객합계']
df
```

```python
res_1 = df.sort_values(by='관광객비율', ascending=False)
res_1
```

```python
res_1 = df.sort_values(by='관광객비율', ascending=False).iloc[1]['사업']
res_2 = df.sort_values(by='관광', ascending=False).iloc[1]['공무']
print(res_1, res_2)
```

```python
# 2) 조건에 맞는 값 찾기
res_1 = df.nlargest(2, '관광객비율').iloc[1]['사업']
res_2 = df.nlargest(2, '관광').iloc[1]['공무']
print(res_1, res_2)
```
