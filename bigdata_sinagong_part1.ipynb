## Section1. 데이터프레임과 시리즈

```python
# 판다스 라이브러리 불러오기
import pandas as pd
```

```python
# 시리즈 만들기 (문자열)
menu = pd.Series(['짜장면', '짬뽕', '탕수육'])
menu
```

```python
# 시리즈 만들기(정수형)
price = pd.Series([4000, 5000, 6000])
price
```

```python
# 데이터 프레임 만들기 pd.DataFrame({"컬럼명":데이터})
pd.DataFrame({'menu': menu,
              'price': price})
```

```python
# 데이터 프레임 만들기 pd.DataFrame({"컬럼명":데이터})
df = pd.DataFrame({
    'menu': ['짜장면', '짬뽕', '탕수육'],
    'price': [4000, 5000, 6000],
    'category': ['중식', '중식', '한식']
})
df
```

```python
# 시리즈 선택 (가격)
df['price']
```

```python
# 데이터프레임 선택 (가격)
df[['price']]
```

```python
df[['menu', 'price']]
```

```python
# 데이터프레임 선택 (가격)
cols = ['menu', 'price']
df[cols]
```

```python
# 데이터 프레임 type 확인
print("df          : ", type(df))
print("df['가격']   : ", type(df['price']))
print("df[['가격']] : ", type(df[['price']]))
```
