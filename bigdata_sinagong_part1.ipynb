```python
# 2개 이상 일때 (OR)
df[cond1 | cond2]
```

```python
# 문자열과 숫자 조건 (AND)
cond1 = df['원산지'] == '과테말라'
cond2 = df['가격'] <= 5000
df[cond1 & cond2]
```
